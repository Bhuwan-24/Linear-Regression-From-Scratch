```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split as tt
import matplotlib.pyplot as plt
import random
```

```python
class Linear_Regression:
    
    def __init__(self,batch_size=10,Learning_Rate=0.01):
        self.coef=None
        self.inter=None
        self.lr=Learning_Rate
        self.batch_size=batch_size

    def fit(self,X,Y):
        X=np.array(X)
        Y=np.array(Y)
        itr=0
        prev_coef=0
        prev_inter=0
        self.coef=np.ones(X.shape[1])
        self.inter=0

        while True:
            itr+=1
            prev_coef=self.coef.copy()
            prev_inter=self.inter
            for i in range(int(X.shape[0]/self.batch_size)):
                indx=random.sample(range(1,X.shape[0]-1),self.batch_size)
                y_pred=np.dot(X.iloc[indx],self.coef)+self.inter
                inter_grad=-2*np.mean(Y.iloc[indx]-y_pred)
                self.inter=self.inter-self.lr*inter_grad
                coef_grad=-2*np.dot(X.iloc[indx].T,(Y.iloc[indx]-y_pred))/self.batch_size
                self.coef=self.coef-self.lr*coef_grad
            if np.all(abs(self.coef-prev_coef)<0.01) and abs(self.inter-prev_inter)<0.01:
                break
            if itr>100:
                break

    def predict(self,x):
        return np.dot(x,self.coef)+self.inter

        

```
