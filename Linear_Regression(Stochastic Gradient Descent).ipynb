```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split as tt
import random
```

```python
class Linear_Regression:
    def __init__(self,LearningRate=0.01):
        self.inter=None
        self.coef=None
        self.lr=LearningRate

    def fit(self,X,Y):
        prev_coef=0
        prev_inter=0
        self.coef=np.ones(X.shape[1])
        self.inter=0
        itr=0

        while True:
            itr+=1
            prev_coef=self.coef.copy()
            prev_inter=self.inter
            for i in range(X.shape[0]):
                indx=random.randint(0,X.shape[0]-1)
                y_pred=np.dot(X[indx],self.coef)+self.inter
                inter_grad=-2*(Y[indx]-y_pred)
                self.inter=self.inter-self.lr*inter_grad
                coef_grad=-2*np.dot((Y[indx]-y_pred),X[indx])
                self.coef=self.coef-self.lr*coef_grad

            if np.all(abs(prev_coef-self.coef)<0.1) and abs(prev_inter-self.inter)<0.1:
                break
            if itr>500:
                break

    def predict(self,x):
        return np.dot(x,self.coef)+self.inter
```
