Linear Regression by Closed Forem method

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split as tts
```

Model Class

```python
class LinearRegression:
    def __init__(self):
        self.m=0
        self.b=0
    def fit(self,X,Y):
        self.X=X
        self.Y=Y
        num=np.sum((X - X.mean()) * (Y - Y.mean()))
        den=np.sum((X-X.mean())**2)
        
        self.m=num/den
        self.b=Y.mean()-(self.m*X.mean())
    def predict(self,x):
        return self.m*x+self.b
```

Data Processing

```python
df=pd.read_csv('insurance.csv')
X,x,Y,y=tts(df['age'],df['charges'],test_size=0.2,random_state=2)
```

Training Model and testing Model

```python
model=LinearRegression()
model.fit(X,Y)
model.predict(x)
```

Visualization

```python
plt.plot(X,model.predict(X), color='red')
plt.scatter(X,Y)
plt.show()
```
