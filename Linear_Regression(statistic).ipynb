```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
class LinearRegression:
    def __init__(self):
        self.m=0
        self.b=0
    def fit(self,X,Y):
        self.num=0
        self.den=0
        for i in range(X.shape[0]):
            self.num=(X[i]-X.mean())*(Y[i]-Y.mean)
            self.den=(X[i]-X.mean())*(X[i]-X.mean())
        self.m=self.num/self.den
        self.b=Y-(self.m*X)
    def predict(self,x):
        return self.m*x+self.b
```

```python
df=pd.read_csv('data.csv')
X=df['Hours']
Y=df['marks']
```

```python
model=LinearRegression()
model.fit(X,Y)
```
