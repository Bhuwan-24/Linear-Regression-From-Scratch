```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
class Linear_Regression:
    
    def __init__(self,Learning_Rate=0.001):
        self.lr=Learning_Rate
        self.m=None
        self.b=None

    def fit(self,X,Y):
        self.m=np.ones(X.shape[1])
        self.b=0
        old_m=0
        old_b=0
        i=0
        while True:
            i+=1
            old_m=self.m.copy()
            old_b=self.b
            y_pred=np.dot(X,self.m)+self.b
            del_b=-2*np.mean(Y-y_pred)
            self.b=self.b-self.lr*del_b
            del_m=-2*np.dot(X.T,(Y-y_pred))/X.shape[0]
            self.m=self.m-self.lr*del_m
            if np.all(abs(old_m-self.m))<0.001 and abs(old_b-self.b)<0.01:
                break
            if(i>1000):
                break
    def predict(self,X):
        return np.dot(X,self.m)+self.b
            



```
