```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split as tt
```

```python
class Linear_Regression:

    def __init__(self,LearningRate=0.01):
        self.coefficient=None
        self.intercept=None
        self.lr=LearningRate
        
    def fit(self,X,Y):
        prev_coefficient=0
        prev_intercept=0
        itr=0
        self.coefficient=np.ones(X.shape[1])
        self.intercept=0
        while True:
            itr+=1
            prev_coefficient=self.coefficient.copy()
            prev_intercept=self.intercept

            y_pred=np.dot(X,self.coefficient)+self.intercept
            intercept_grad=-2*np.mean(Y-y_pred)
            self.intercept=self.intercept-self.lr*intercept_grad
            coefficient_grad=-2*np.dot(X.T,(Y-y_pred))/X.shape[0]
            self.coefficient=self.coefficient-self.lr*coefficient_grad

            if np.all(abs(prev_coefficient-self.coefficient)<0.01) and abs(prev_intercept-self.intercept)<0.01:
                break           
            if itr>1000:
                break

    def predict(self,x):
            return np.dot(x,self.coefficient)+self.intercept
```

```python
df=pd.read_csv('insurance.csv')
df.drop(['sex','smoker','region'],axis=1,inplace=True)
Y=df['charges']
X=df.drop('charges',axis=1)
X=(X-X.mean())/X.std()
```

```python
x_train,x_test,y_train,y_test=tt(X,Y,test_size=0.2,random_state=2)
```

```python
lr=Linear_Regression(0.1)
lr.fit(x_train,y_train)
lr.predict(x_test)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```
