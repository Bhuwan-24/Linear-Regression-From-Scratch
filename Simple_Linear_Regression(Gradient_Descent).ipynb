```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Gradient Descent

```python
class Linear_Regression():

    def __init__(self,LearningRate):
        self.m=0
        self.b=0
        self.lr=LearningRate
        self.itr=0
        
    def fit(self,X,Y):

        if X.ndim>1:
            X=X.ravel()
        if Y.ndim>1:
            Y=Y.ravel()
        while True:
            prev_m=self.m
            prev_b=self.b
            slope_m=-2*np.sum(X*(Y-self.m*X-self.b))
            self.m=self.m-self.lr*slope_m
            slope_b=-2*np.sum(Y-self.m*X-self.b)
            self.b=self.b-self.lr*slope_b
            self.itr+=1
            if(abs(prev_m-self.m)<0.0001 and abs(prev_b-self.b)<0.0001):
                break
            if self.itr>10000:
                break
                
    def predict(self,x):
        return self.m*x+self.b
```

Data Loading

```python
df=pd.read_csv('data.csv')
X=df["Hour"]
Y=df["marks"]
```

prediction

```python
model=Linear_Regression(0.001)
model.fit(X,Y)
model.predict(20)
```

Visualization

```python
plt.scatter(X,Y,color='red')
plt.plot(X,model.predict(X))
plt.show()
```

Comparision with Sk learn model

```python
print(model.b,model.m)
lr=LinearRegression()
lr.fit(np.array(X).reshape(-1,1),np.array(Y).reshape(-1,1))
print(lr.coef_,lr.intercept_)
```
