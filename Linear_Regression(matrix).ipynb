```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
class Linear_Regression:
    def __init__(self):
        self.coef=0
        self.inter=0
    def fit(self,x,y):
        x=np.array(x)
        y=np.array(y).reshape(-1,1)
        x=np.insert(x,0,1,axis=1)
        var=np.linalg.pinv(x)@y
        self.inter=var[0]
        self.coef=var[1:]
    def predict(self,X):
        X=np.array(X)
        if X.ndim==1:
            X.reshape(-1,1)
        pred=X@self.coef+self.inter
        return pred
```

```python
df=pd.read_csv('insurance.csv')
df.head()
df.drop(['sex','smoker','region'],axis=1,inplace=True)
x=df.drop('charges',axis=1)
y=df['charges']
```

```python
model=Linear_Regression()
model.fit(x,y)
model.predict([10,21,0])
```
